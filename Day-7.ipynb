```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

print("="*60)
print("QUESTION & ANSWER CHATBOT")
print("="*60)

print("Ask any question.")
print("Type 'exit' to quit.\n")


# -------------------------------------------------
# Knowledge Base
# -------------------------------------------------

knowledge_base = {

    "python":
        "Python is a high-level, interpreted programming language used for AI, web development, automation, and data science.",

    "ai":
        "Artificial Intelligence (AI) enables computers to simulate human intelligence.",

    "machine learning":
        "Machine Learning is a subset of AI that enables systems to learn from data without explicit programming.",

    "deep learning":
        "Deep Learning uses multi-layer neural networks for solving complex problems.",

    "nlp":
        "Natural Language Processing enables computers to understand and process human language.",

    "rag":
        "Retrieval-Augmented Generation combines document retrieval with language models.",

    "chatbot":
        "A chatbot is a software application that interacts with humans using natural language.",

    "tokenization":
        "Tokenization is the process of splitting text into words or sentences.",

    "database":
        "A database is an organized collection of structured information.",

    "cloud":
        "Cloud Computing delivers computing resources over the Internet.",

    "iot":
        "Internet of Things connects physical devices through the Internet.",

    "blockchain":
        "Blockchain is a decentralized digital ledger technology.",

    "cyber security":
        "Cyber Security protects computer systems from digital attacks."
}

# -------------------------------------------------
# Chat Loop
# -------------------------------------------------

while True:

    question = input("\nYou : ").lower()

    if question == "exit":
        print("\nBot : Thank you!")
        break

    tokens = word_tokenize(question)
    found = False

    # Search every keyword in knowledge base

    for key in knowledge_base:

        key_tokens = key.split()

        if all(word in tokens for word in key_tokens):

            print("\nBot :")
            print(knowledge_base[key])

            found = True
            break

    if not found:

        print("\nBot :")
        print("Sorry, I don't have an answer for that question.")

```
